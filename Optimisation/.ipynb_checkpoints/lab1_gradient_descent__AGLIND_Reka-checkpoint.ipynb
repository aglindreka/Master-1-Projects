## Q1


To determine if the function 𝑓(𝑥) is convex, we need to examine the Hessian matrix of 𝑓(𝑥).
A function is convex if its Hessian matrix is positive definite everywhere, if all of its eigenvalues are positive.

#### Derivates

$$𝑓(𝑥) = (𝑥0 − 𝑥1 − 1)^2 + 𝜇/2 * 𝑥0^2$$

$$df/dx0 = 2(𝑥0 − 𝑥1 − 1) + 𝜇𝑥0$$

$$d𝑓/d𝑥1 = −2(𝑥0 − 𝑥1 − 1)$$

$$d𝑓/d𝑥0𝑥0 = 2 + 𝜇$$

$$d𝑓/d𝑥0𝑥1 = −2$$

$$d𝑓/d𝑥1𝑥0 = −2$$

$$d𝑓/d𝑥1𝑥1 = 2$$

The hessian matrix is: 

\
\begin{pmatrix}
2+𝜇 & -2 \\
-2 & 2
\end{pmatrix}
\

To determine the eigenvalues of H(𝑓), we need to solve the characteristic equation:

det(H(𝑓) - λI) = 0

where I is the identity matrix and λ is the eigenvalue.

Expanding the determinant, we get:

(2 + 𝜇 - λ) (2 - λ) - (-2) (-2) = 0



Like we see in the picture the eigenvalues are positive if 𝜇 >= 0
 Therefore, both eigenvalues of H(𝑓) are positive for all values of 𝑥, and hence, 𝑓(𝑥) is a convex function.

The minimum eigenvalue of this matrix is 𝜇, so 𝑓(𝑥) is strongly convex if ``𝜇 > 0``.

Therefore, 𝑓(𝑥) is strongly convex for all 𝑥 ∈ ℝ² when 𝜇 > 0.
So to be strongly convex the 𝜇 should be > 0.

## Q2

To find the unique minimizer of the function 𝑓(𝑥)  when 𝜇 > 0, we need to find the critical points of 𝑓(𝑥) by setting its gradient equal to zero


(-1/ 𝜇, -1/ 𝜇 + 1)

````python
def f2D_convex(x, mu=1.0):
    return (x[0]-x[1]-1)**2+(mu/2)*x[0]**2
````

**Q6**. Is it a convex function? Describe the minima of $f$.<br>


### Q6. ANS

To show this, we need to do the same as in the first exercise, we need to check if the Hessian matrix of the function is positive semidefinite for all values of x0 and x1. The Hessian matrix is the matrix of second partial derivatives of the function



Taking the second partial derivatives of f(x0, x1), we get:

$$d²f/dx0² = -cos(x0)$$
$$d²f/dx1² = -sin(x1)$$
$$d²f/dx0dx1 = 0$$
$$d²f/dx0dx1 = 0$$

Therefore, the Hessian matrix of $$f(x0, x1)$$ is:
\
\begin{pmatrix}
-cos(x0) & 0 \\
0 & -sin(x1)
\end{pmatrix}
\


Since the eigenvalues of H(f) are -cos(x0) and -sin(x1), which can be negative for certain values of x0 and x1, the Hessian matrix is not positive semidefinite for all values of x0 and x1. This implies that the function f(x0, x1) is not convex.

#### Describe the minima of $f$

To find the minima of the given function we can take the partial derivatives of the function with respect to x0 and x1, set them to zero, and solve for x0 and .

$$-sin(x0) = 0$$
$$cos(x1) = 0$$

Solving for x0 and x1, we get:

$$x0 = nπ, $$where n is an integer
$$x1 = (2m + 1)π/2$$, where m is an integer

Therefore, the critical points of f(x0, x1) occurs at the points $$(nπ, (2m + 1)π/2)$$, where n and m are integers.

To determine if these critical points are minima, we need to check the second derivative test. The Hessian matrix of f(x0, x1) is:
Taking the second partial derivatives of f(x0, x1), we get:

\
\begin{pmatrix}
-cos(x0) & 0 \\
0 & -sin(x1)
\end{pmatrix}
\


At the critical points, we have:

$$d²f/dx0² = -cos(nπ) = (-1)^n$$
$$d²f/dx1² = -sin((2m + 1)π/2) = (-1)^m$$

Therefore, the Hessian matrix at the critical points is:


\begin{pmatrix}
(-1)^n & 0 \\
0 & (-1)^m
\end{pmatrix}


The eigenvalues of H(f) are (-1)^n and (-1)^m, which can be positive or negative depending on the values of n and m. Therefore, the critical points can be either a minimum, maximum, or a saddle point depending on the values of n and m.

In general, there are an infinite number of critical points, and the minima of f(x, y) occur at the critical points where both eigenvalues of the Hessian matrix are positive.

## Gradient descent procedure

We recall that a gradient step takes the form
$$ x^{(t+1)} = x^{(t)} - \eta^{(t)} \nabla f(x^{(t)}) $$
with some initialization $x^{(0)}$ and $\eta^{(t)}$ a learning rate policy.

**Q1**. Implement a function `gradient_descent_tmp` taking as argument `grad_f`, `eta`, `x0` and `max_iter`. `grad_f` will take as input an optimization variable `x`, `eta` is the learning policy taking as arguments `x` and `t` the current iteration, `x0` is the initialization and `max_iter` a maximal number of iterations. As an example, a constant learning rate policy can defined as
```python
def constant_eta(x, t):
    return 0.01
```

``What is gradient descent?`` 
Gradient descent is a way to find the minimum value of a function by taking small steps in the direction of the negative gradient of the function at a given point.


````python
def gradient_descent_tmp(grad_f, eta, x0, max_iter):
    x = x0
    for t in range(max_iter):
        x=x-eta(x,t)*grad_f(x) #x is vector, eta is value
    return x
````

**Q2**. Implement a function `grad_f2D_convex` returning the gradient of the convex function defined in Section 1.

````python
def grad_f2D_convex(x, mu=1.0):
    return np.array([
        2*(x[0]-x[1]-1)+mu*x[0],
        -2*(x[0]-x[1]-1)
    ])
````

#### Q2

The gradient of the function is given by:

grad_f2D_convex:

\
\begin{pmatrix}
df/dx0 & df/dx1
\end{pmatrix}


where df/dx0 is the partial derivative of f2D_convex with respect to x0, and df/dx1 is the partial derivative of f2D_convex with respect to x0.

The derivates I have calculated even in the previous exercises.

Therefore, the gradient of the function f2D_convex(x, mu=1.0) is:

grad_f2D_convex:

\
\begin{pmatrix}
2*(x0-x1-1)+mu*x0 & -2*(x0-x1-1)
\end{pmatrix}

**Q3**. Compute the Lipschitz constant of $\nabla f(x)$ and deduce a learning rate policy `lr_policy_f2D-convex(x,t)`. Discuss the difference between $\mu = 0$ and $\mu \neq 0$.

````python
def hessian_f2D_convex(x, mu=1.0):
    return np.array([
        [2+mu, -2],
        [-2, 2]
    ])


def lr_policy_f2D_convex(x, t):
    return 1/np.linalg.norm(hessian_f2D_convex(x),2)
````

The first function, hessian_f2D_convex(x, mu=1.0), computes the Hessian matrix of a two-dimensional convex function with respect to the input variable x. 

The second function, lr_policy_f2D_convex(x, t), computes the learning rate for gradient descent based on the Hessian matrix of the same two-dimensional convex function. The learning rate is computed as the inverse of the L2 norm of the Hessian matrix.

hessian_f2D_convex(x, mu) function provided, the parameter mu determines the strength of the off-diagonal elements in the Hessian matrix. Specifically, the Hessian matrix is given by:

    
\begin{pmatrix}
2+mu & -2 \\
-2 & 2
\end{pmatrix}
    
mu = 0, Hessian matrix has eigenvalues of 4 and 0, which means that the function has a minimum at the origin (0,0) and is flat in one direction.


mu is different from 0, it modifies the curvature of the function, the location of the minimum point, and the convergence speed of the optimization algorithm. These effects can be either positive or negative, depending on the specific value of mu and the optimization problem at hand.

**Q5**. Implement a function `gradient_descent` taking as arguments `grad_f`, `eta`, `x0`, `max_iter` (defined as in *2.Q1*), `return_all` a boolean indicating if we keep in memory all iterates and `callback` a callable that is called at each iteration if not `None`. If `return_all` and `callback` are `None`, it should behave like `gradient_descent_tmp`. Otherwise, it should return a `dict` containing the final iterate, and eventually all the iterates and outputs of the callback. The function may look like the following stub
```python
def gradient_descent(grad_f, eta, x0, max_iter, return_all=False, callback=None):
    x = x0
    fxs = []

```

````python
def gradient_descent(grad_f, eta, x0, max_iter, return_all=False, callback=None):
    x = x0
    fxs = None #return numpy.ndarray not python list
    xs = None
    
    if callback:
        fx0=callback(x0)
        fxs=np.zeros((max_iter,*fx0.shape)) #*fx0 means there are number of parameters: fx0[0], fx0[1]...
        fxs[0,...]=fx0 #ellipsis operator ... even fxs[0] is correct its just that fx[0] is array in itself
        
    if return_all:
        xs=np.zeros((max_iter,*x0.shape))
        xs[0]=x0
        
    for t in range(max_iter):
        x=x-eta(x,t)*grad_f(x)
        if return_all:
            xs[t]=x
        if callback:
            fxs[t]=callback(x)
        
    # Do something
    if not callback and not return_all:
        return x
    else:
        return {
            'res': x,
            'callbacks': fxs,
            'iterates': xs,
        }
````

To perform our classification task, we will turn to the multinomial logistic regression, or softmax classification.
The objective function here will be seek a weight matrix $W \in \mathbb{R}^{K \times p}$

$$ f(W) = -\frac1 n\sum_{i=1}^n \sum_{k=0}^{K-1} [y_i = k] \log(\hat p(y_i = k | X_i)) + \lambda \| W \|_2^2, $$
where
$$ \hat p(y_i = k | X_i) = \frac{\exp(X_i W_k)}{\sum_{l=0}^{K-1}\exp(X_i W_l)} . $$

$$ f(W) = -\frac1 n\sum_{i=1}^n (sum-over-all-training-examples) \sum_{k=0}^{K-1} [y_i = k] (sum-over -all-classes)\log(\hat p(y_i = k | X_i)) + \lambda \| W \|_2^2, $$

``The softmax function`` is commonly used in machine learning as the last layer of a neural network to convert the output of the network into a probability distribution over a set of classes. The class with the highest probability is often chosen as the predicted class for a given input.
$$ \hat p(y_i = k | X_i) = \frac{\exp(X_i W_k)}{\sum_{l=0}^{K-1}\exp(X_i W_l)} . $$

**Numerical stability of the softmax**. When computing the softmax of a vector, one has to be careful of the potential numerical errors due to the large values of $\exp$ when the argument is above 0.
This can result in overflow errors or loss of precision in the calculation.

````python
scores = np.array([433, 23, 766])
proba = np.exp(scores) / np.sum(np.exp(scores))
proba
````

Clearly here, we would expect to obtain `[0., 0., 1.]`.
Remark that for any $c > 0$,
$$
\sigma(z)_i
=
\frac{\exp(z_i)}{\sum_{i=1}^p \exp{(z_j)}}
=
\frac{c\exp(z_i)}{c\sum_{i=1}^p \exp{(z_j)}}
=
\frac{\exp(z_i + \log c)}{\sum_{i=1}^p \exp{(z_j + \log c)}}
$$
A typical choice of $c$ is given by the solution of $\log c = - \max_j z_j$.

````python
scores = np.array([433, 23, 766])
scores -= np.max(scores)
proba = np.exp(scores) / np.sum(np.exp(scores))
proba
````

``The given code is using the softmax function to convert the given scores into probabilities.``

**Q2**. Implement the `softmax_loss` function below. Take care to the potential numerical errors thanks to the trick above. _Hint_: remember that `numpy.arange(n)` build a ndarray containing all integer from 0 to `n` (included).

````python
x=np.random.randn(4,5)
np.sum(x,axis=1,keepdims=True).shape
````

````python
def softmax_loss(W, X, y, reg):
    """
    Softmax loss function

    Inputs:
    - W: ndarray (n_features, K) containing weights.
    - X: ndarray (n, n_features) containing data.
    - y: ndarray (n,) containing training labels
    - reg: (float) regularization

    Returns the softmax loss (with regularization)
    """
    n,_=X.shape
    raw_scores=X@W
    
    #Numerical stable softmax
    proba=np.exp(raw_scores-raw_scores.max())
    proba/=np.sum(proba,axis=1,keepdims=True) #keepdims for broadcasting
    
    #Cross-entropy
    loss=-np.sum(np.log(proba[np.arange(n),y]))/n #this is justof size n 
    #add regularization
    loss+=reg*np.sum(W**2)
    return loss
````

**Q5**. Implement the gradient of the softmax loss as a `softmax_grad` function below. _Hint_: This function will be _very_ close to `softmax_loss`. Note that it would be more adequate to write a unique function returning _both_ the loss and the gradient.

````python
def softmax_grad(W, X, y, reg):
    n,_=X.shape
    raw_scores=X@W
    
    #Numerical stable softmax
    proba=np.exp(raw_scores-raw_scores.max())
    proba/=np.sum(proba,axis=1,keepdims=True)
    
    #update prob vector
    proba[np.arange(n),y]-=1
    
    dW=X.T@proba/n
    dW+=2*reg*W
    return dW
````

``The gradient of the loss function is used in gradient descent or other optimization algorithms to update the weight matrix W during training, with the goal of minimizing the loss function and improving the model's accuracy.``
